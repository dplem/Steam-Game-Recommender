```python
steam_df = steam_df.stack().str.replace(',',' ').unstack()
steam_df.head()
```

```python
steam_df['game'] = steam_df['game'].str.lower()
steam_df['popular_tags'] = steam_df['popular_tags'].str.lower()
steam_df['game_details'] = steam_df['game_details'].str.lower()
steam_df['genre'] = steam_df['genre'].str.lower()
steam_df.head()
```

```python
steam_df['soup'] = steam_df[steam_df.columns[1:]].apply(
    lambda x: ' '.join(x.dropna().astype(str)),
    axis=1
)
steam_df['soup'].head()
```

```python
#Import TfIdfVectorizer from scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#Define a TF-IDF Vectorizer Object. Remove all english stop words such as 'the', 'a'
tfidf = TfidfVectorizer(stop_words='english')
```

```python
soup = steam_df['soup']
soup.head()
```

```python
#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(soup).toarray()
tfidf_matrix
```

```python
#Array mapping from feature integer indices to feature name.
tfidf.get_feature_names()
```

```python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel
```

```python
# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
cosine_sim.shape
```

```python
cosine_sim[1]
```

```python
#Construct a reverse map of indices and game titles
indices = pd.Series(steam_df.index, index=steam_df['game']).drop_duplicates()
```

```python
indices[:10]
```

```python
# Function that takes in a game title as input and outputs most similar games
def get_recommendations(title, cosine_sim=cosine_sim):
    # Get the index of the game that matches the title
    idx = indices[title]

    # Get the pairwsie similarity scores of all games with that game
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the games based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar games
    sim_scores = sim_scores[1:11]

    # Get the game indices
    game_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar games
    return steam_df['game'].iloc[game_indices]
```

```python
get_recommendations('doom')
```

```python
get_recommendations('eve online')
```
